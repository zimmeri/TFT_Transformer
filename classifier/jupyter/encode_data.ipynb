```python
import json
import pandas as pd
import category_encoders as ce
from sklearn.preprocessing import LabelEncoder
pd.set_option('display.max_columns', None)
```

```python
match_data = pd.read_csv('../data/dataset_NA_BR_EUW_KR_section1.csv', delimiter='\t') #read in match data
match_data = match_data.replace(0, '0') #convert all 0s to chars to standardize dataset
```

```python
trait_encoder = ce.BinaryEncoder(cols=['Trait 1', 'Trait 2', 'Trait 3', 'Trait 4', 'Trait 5', 'Trait 6', 'Trait 7', 'Trait 8']) #binary encode the traits
trait_encoded_df = trait_encoder.fit_transform(match_data)
trait_encoded_df
```

```python
champion_encoder = ce.BinaryEncoder(cols=['Champion 1', 'Champion 2', 'Champion 3','Champion 4', 'Champion 5', 'Champion 6','Champion 7', 'Champion 8', 'Champion 9', 'Champion 10', 'Champion 11']) #binary encode the units
champion_encoded_df = champion_encoder.fit_transform(trait_encoded_df)
champion_encoded_df
```

```python
item_econder = ce.BinaryEncoder(cols=['Item 1 1', 'Item 1 2', 'Item 1 3',
                                    'Item 2 1', 'Item 2 2', 'Item 2 3',
                                    'Item 3 1', 'Item 3 2', 'Item 3 3',
                                    'Item 4 1', 'Item 4 2', 'Item 4 3',
                                    'Item 5 1', 'Item 5 2', 'Item 5 3',
                                    'Item 6 1', 'Item 6 2', 'Item 6 3',
                                    'Item 7 1', 'Item 7 2', 'Item 7 3',
                                    'Item 8 1', 'Item 8 2', 'Item 8 3',
                                    'Item 9 1', 'Item 9 2', 'Item 9 3',
                                    'Item 10 1', 'Item 10 2', 'Item 10 3',
                                    'Item 11 1', 'Item 11 2', 'Item 11 3']) #binary encode the items
item_encoded_df = item_econder.fit_transform(champion_encoded_df)
item_encoded_df
```

```python
item_encoded_df.to_csv('../data/test_dataset_encoded.csv', sep='\t', encoding='utf-8', index=False, header=True) #save the data
```
